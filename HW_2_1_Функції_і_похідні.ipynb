### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.symbols('x')
f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)
print("Похідна лінійної функції 10x-5:", derivative_f_lin)
```

```python
xs = np.linspace(0, 10, 100)
f_lin = 10 * xs - 5
df_lin = 10
df_lin_vals = np.full_like(xs, df_lin)
plt.figure(figsize=(10, 5))
plt.plot(xs, f_lin, label='f(x) = 10*x-5', color='blue')
plt.plot(xs, df_lin_vals, label="f'(x) = 10", color='green')

plt.title("Графік функції 10x-5 та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()

plt.show()
```

зі збільшенням кількості годин витрачених на навчання збільшується оцінка. Оцінка збільшується лінійно. Похідна є сталою, що вказує на рівномірний темп покращення оцінки з кожною годиною навчання.
